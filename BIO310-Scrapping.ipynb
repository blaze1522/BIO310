```python
import pandas as pd
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.common.exceptions import StaleElementReferenceException
import time
```

```python
timeout = 15
driver = webdriver.Firefox()
class_title = 'js-article-title'
class_type = 'js-article-subtype'
```

Here we are going to define a function which takes the volume and issue nuber and gives then web scraps to get all the required titles and the type of publication published in that issue.

Note: For slower internet connection increase sleep time on Line:10

```python
def find_title(vol, issue):
    driver.refresh()
    url = 'https://www.sciencedirect.com/journal/neuron/vol/'+vol+'/issue/'+issue
    driver.get(url)

    # Wait for the required class of elements in this case titles to appear.
    WebDriverWait(driver, timeout).until(
        EC.visibility_of_element_located((By.CLASS_NAME, class_title)))
    print("Found")
    time.sleep(2)
    
    # Pick all the elements with required class
    titles = driver.find_elements_by_class_name(class_title)
    types = driver.find_elements_by_class_name(class_type)
    print('Picked')
    
    title_list = [x.text for x in titles]
    type_list = [x.text for x in types]
    print('List made')

    article_table = zip([int(vol)]*len(type_list),[int(issue)]*len(type_list),range(len(type_list)), type_list, title_list)
    
    return list(article_table)
```

I am assigned the year 2014 so I need the data for volumes in 2013, 2012 which are 73 to 80. Every volume has 6 issues.

```python
data = []
for vol in range(73,81):
    for issue in range(1,7):
        print('Volume {} Issue {}'.format(vol,issue))
        titles = find_title(str(vol), str(issue))
        data.extend(titles)

df = pd.DataFrame(data, columns=['vol', 'issue', 'index', 'type', 'title'])
df.to_csv('/home/goirik/Documents/IISER/Sem 5/BIO310 BioStats/Assignment/data.csv')

print('All Done')
```
